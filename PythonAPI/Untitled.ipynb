```python
import pandas as pd

# Load existing dataset
df = pd.read_csv("gameplay_data.csv")

# Drop rows where all button outputs are zero (no action)
button_cols = ["up", "down", "left", "right", "Y", "B", "X", "A", "L", "R"]
df = df.drop_duplicates()
df_cleaned = df[~(df[button_cols].sum(axis=1) == 0)]

# Optional: Reset index
df_cleaned.reset_index(drop=True, inplace=True)


# # Save cleaned version
df_cleaned.to_csv("gameplay_data_cleaned.csv", index=False)

# print(f"✅ Cleaned data saved. Rows before: {len(df)}, after cleaning: {len(df_cleaned)}")
df_cleaned
```
